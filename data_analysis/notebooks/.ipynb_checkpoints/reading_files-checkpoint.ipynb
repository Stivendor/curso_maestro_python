```python
# if we want to quit the header name we do the next thing 
df6 = pd.read_csv("../data/supermarkets/data.csv", header=None)
df6
```

```python
# if we want add the name of a data frame header we do this
# this show double header because in the original file it was the header
df6.columns = ["ID", "Address", "City", "State", "Country", "Name", "Employees"]
df6
```

```python
# to set the correct value of an index, in this case the column ID, we do this
df7 = pd.read_csv("../data/supermarkets/supermarkets.csv",index_col="ID")
df7 
```

```python
# if we want to filter one portion of the info, we do this with the .loc
df7.loc[2:3,"Country":"Employees"]
```

```python
# if we want to get only one value we should do this
df7.loc[1,"Address"]
```

```python
# if we want to get all values of a column we should do this
list(df7.loc[:,"Address"])
```

```python
# access by indexing using iloc
df7.iloc[0:2,3:]
```

```python
# delete the column "City"
df7.drop("Address", axis=1, inplace=True)
df7
```

```python
# delete the row with index 1
df7.drop(1, axis=0, inplace=True)
df7
```

```python
# adding new rows
df7["Continent"]=df7.shape[0]*["North America"]
```
